```python
x = np.arange(0, 11)
```

```python
x
```

```python
y = np.arange(0,11) ** 2
```

```python
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, color = 'Red')
plt.xlabel('My X-values')

plt.show()
plt.ylabel('My Y-values')
```

```python
plt.plot(x, y, color = 'Red', marker = 'o')
plt.xlabel('My X-values')
plt.ylabel('My Y-values')
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot(x,y)
plt.title('XvsY')

plt.subplot(1,3,2)
plt.plot(y,x, c = 'r')
plt.title('YvsX')

plt.subplot(1,3,3)
plt.plot(x,x, c = 'g')
plt.title('XvsX')
plt.show()
```

```python
# Object oriented

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #lenght, breadth, width, height

axes.plot(x, y, color = 'y')
axes.set_title('X vs Y')
axes.set_xlabel('X values')

plt.show()
```

```python
fig = plt.figure(figsize = (10,6))
axes1 = fig.add_axes([0.1, 0.2, 0.8, 0.8]) #lenght, breadth, width, height

axes1.plot(x, y, color = 'y')
axes1.set_title('X vs Y')
axes1.set_xlabel('X values')

axes2 = fig.add_axes([0.2, 0.6, 0.3, 0.3])
axes2.plot(y, x, c = 'r')
axes2.set_title('Inner Plot')

plt.show()
```

```python
fig, axes = plt.subplots(1,2)

axes[0].plot(x, y)
axes[0].set_title('First')
axes[1].plot(y, x)
axes[1].set_title('Second')

plt.show()
```

```python
fig = plt.figure(figsize = (10, 6))

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, marker = 'o', c = 'r', mfc = 'yellow', markeredgecolor = 'blue',
         label = 'First', lw = 3, alpha = 0.2)
axes.plot(x, x, marker = 'x', c = 'g', label = 'Second', ls = '--')
axes.legend(loc = 2)

plt.show()
```
